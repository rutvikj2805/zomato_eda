```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

df = pd.read_csv('./Zomatodataset/zomato.csv', encoding='latin-1')
df.head()
```

```python
country_code = pd.read_excel('./Zomatodataset/Country-Code.xlsx')
country_code.head()
```

```python
#merging both similar data frames

final_df = pd.merge(df, country_code, on='Country Code', how='left')
final_df.head()

#if we have two datasets and if we can merge both into one df then we must do it
```

```python
df.dtypes #to check data types
```

```python
final_df.columns
```

```python
country_names = final_df.Country.value_counts().index
```

```python
country_values = final_df.Country.value_counts().values
```

```python
#plot a pie chart to see the top 3 countries where the transaction are most frequent

plt.pie(country_values[:3], labels=country_names[:3], autopct='%1.2f%%')
```

```python
final_df.columns
```

```python
rating_df = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns = {0: 'Rating count'})
rating_df
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x = 'Aggregate rating', y = 'Rating count', hue='Rating color', palette=['White', 'Red', 'Orange', 'Yellow', 'Green', 'Green'], data = rating_df)
```

```python
## Count Plot(only applies on categorical values)

sns.countplot(x='Rating color', hue='Rating color', palette=['White', 'Red', 'Orange', 'Yellow', 'Green', 'Green'], data = rating_df)
```
